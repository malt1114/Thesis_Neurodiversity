```python
from nilearn import image as nimg
import numpy as np
import pandas as pd
```

```python
img_small = nimg.load_img('data.nosync/Yeo2011_7Networks_MNI152_FreeSurferConformed1mm_LiberalMask.nii.gz')
img_small = img_small.get_fdata()

img_large = nimg.load_img('data.nosync/Yeo2011_17Networks_MNI152_FreeSurferConformed1mm_LiberalMask.nii.gz')
img_large = img_large.get_fdata()

size_of_roi = {}
for i in np.unique(img_small):
    size_of_roi[i] = img_small[img_small == i].shape[0]

overlab_data = []

for i in np.unique(img_large):
    temp = {'ROI': i}
    for key, value in size_of_roi.items():
        temp[f"ROI_{key}"] = (img_small[(img_small == key) & (img_large == i)].shape[0]/value)*100
    overlab_data.append(temp)

data = pd.DataFrame(overlab_data)

data[data>5]
```
